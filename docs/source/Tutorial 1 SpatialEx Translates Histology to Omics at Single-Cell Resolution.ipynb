````python
h5_path = './datasets/Human_Breast_Cancer_Rep2/cell_feature_matrix.h5'
obs_path = './datasets/Human_Breast_Cancer_Rep2/cells.csv'
img_path = './datasets/Human_Breast_Cancer_Rep2/Xenium_FFPE_Human_Breast_Cancer_Rep2_he_image.ome.tif'
trans_mtx_path = './datasets/Human_Breast_Cancer_Rep2/Xenium_FFPE_Human_Breast_Cancer_Rep2_he_imagealignment.csv'
ref = pp.Read_Xenium(h5_path, obs_path)    
ref = pp.Preprocess_adata(ref, cell_mRNA_cutoff=0)      

obs = sc.read_h5ad('./datasets/Human_Breast_Cancer_Rep2/Human_Breast_Cancer_Rep2_uni_resolution64_full.h5ad')

# ours_313_panelA2 is the predicted gene expression within the sequencing area 
x_prime = pd.read_csv('./output/313_gene/ours_313_panelA2.csv', index_col=0)
ref = ref[x_prime.index.astype(str)]
obs = obs[x_prime.index.astype(str)]
ref.obs = obs.obs.copy()
ref.obsm['image_coor'] = obs.obsm['image_coor']
x_prime = np.maximum(x_prime.values, 0)
adata = ref.copy()
adata.X = x_prime                                     # the prediction of squencing area

# Rep2_out_HA is the predicted gene expression for the whole histology
panelA = np.load('./output/313_gene/Rep2_out_HA.npy')
panelA = np.maximum(panelA, 0)
obs = pd.read_csv('./datasets/Human_Breast_Cancer_Rep2/HBRC_Rep2_cell_coor.csv', index_col=0)
obs.shape                                             # the prediction of out of squencing area within panelA
ref
````

determine the sequencing area

````python
col_min, col_max = ref.obsm['image_coor'][:, 1].min(), ref.obsm['image_coor'][:, 1].max()
row_min, row_max = ref.obsm['image_coor'][:, 0].min(), ref.obsm['image_coor'][:, 0].max()
selection = (obs['image_row'] > row_min) & (obs['image_row'] < row_max) & (obs['image_col'] > col_min) & (obs['image_col'] < col_max)
obs_inner = obs[selection]
var_names = adata.var_names
````

#### Draw the gene *EPCAM*

````python
# draw the predicted spatial expression within the sequencing area
value = adata[:, 'EPCAM'].X
vmax = ref[:, 'EPCAM'].X.max()

x, y = ref.obsm['image_coor'][:, 0], ref.obsm['image_coor'][:, 1]
plt.scatter(x, y, c=value, vmax=vmax-0.8, s=0.01)                    
# draw red box
x = [row_min, row_min, row_max, row_max, row_min]
y = [col_min, col_max, col_max, col_min, col_min]
plt.plot(x, y, color='red', linestyle='--')
# draw the predicted spatial expression out of the sequencing area
gene_idx = np.where(var_names=='EPCAM')[0].item()
value = panelA[:, gene_idx]
x, y = obs['image_col'], obs['image_row']
plt.scatter(y[~selection], x[~selection], c=value[~selection], s=0.01, vmin=0, vmax=vmax-0.8)

plt.title('EPCAM')
plt.ylim((0,x.max()))
plt.xlim((0,y.max()))
plt.axis('off')
ax = plt.gca()
ax.set_aspect(1)
````

#### Draw the *ESR1* gene

````python
# draw the predicted spatial expression within the sequencing area
value = adata[:, 'ESR1'].X
vmax = ref[:, 'ESR1'].X.max()
# vmax = value.max()
x, y = ref.obsm['image_coor'][:, 0], ref.obsm['image_coor'][:, 1]
plt.scatter(x, y, c=value, vmax=vmax-0.8, s=0.01)                    
# draw red box
x = [row_min, row_min, row_max, row_max, row_min]
y = [col_min, col_max, col_max, col_min, col_min]
plt.plot(x, y, color='red', linestyle='--')
# draw the predicted spatial expression out of the sequencing area
gene_idx = np.where(var_names=='ESR1')[0].item()
value = panelA[:, gene_idx]
x, y = obs['image_col'], obs['image_row']
plt.scatter(y[~selection], x[~selection], c=value[~selection], s=0.01, vmin=0, vmax=vmax-0.8)

plt.title('ESR1')
plt.ylim((0,x.max()))
plt.xlim((0,y.max()))
plt.axis('off')
ax = plt.gca()
ax.set_aspect(1)
````

#### Draw the PGR gene

````python
# draw the predicted spatial expression within the sequencing area
value = adata[:, 'PGR'].X
vmax = ref[:, 'PGR'].X.max()
# vmax = value.max()
x, y = ref.obsm['image_coor'][:, 0], ref.obsm['image_coor'][:, 1]
plt.scatter(x, y, c=value, vmax=vmax-0.8, s=0.01)                    
# draw red box
x = [row_min, row_min, row_max, row_max, row_min]
y = [col_min, col_max, col_max, col_min, col_min]
plt.plot(x, y, color='red', linestyle='--')
# draw the predicted spatial expression out of the sequencing area
gene_idx = np.where(var_names=='PGR')[0].item()
value = panelA[:, gene_idx]
x, y = obs['image_col'], obs['image_row']
plt.scatter(y[~selection], x[~selection], c=value[~selection], s=0.01, vmin=0, vmax=vmax-0.8)

plt.title('PGR')
plt.ylim((0,x.max()))
plt.xlim((0,y.max()))
plt.axis('off')
ax = plt.gca()
ax.set_aspect(1)
````

download the predicted cell type 'Out_celltype.npy' from [there]() and the cell coordinate 'HBRC_Rep2_cell_coor.csv' from [there]()

````python
celltype = np.load('./output/313_gene/Out_celltype.npy')
obs = pd.read_csv('./datasets/Human_Breast_Cancer_Rep2/HBRC_Rep2_cell_coor.csv', index_col=0)
x, y = obs['image_col'], obs['image_row']
ct_list = list(color_dict.keys())
for i in range(len(ct_list)):
    ct = ct_list[i]
    selection = (celltype == ct)
    plt.scatter(y[selection], x[selection], c=color_dict[ct], s=0.01, label=ct)
plt.legend(bbox_to_anchor=(1, 0), loc=3, borderaxespad=0, markerscale=100)
ax = plt.gca()
ax.set_aspect(1)
plt.axis('off')
````
